```python
df = df.dropna()
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
# Check missing values
print(df.isnull().sum())

# Fill numerical missing values with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())


# Remove duplicates
df = df.drop_duplicates()

# Convert data types if needed
df['Launch_Year'] = df['Launch_Year'].astype(int)

# Final check
print(df.isnull().sum())
```

```python
# Statistical summary
print(df.describe())

# Unique values in important columns
print(df['Category'].value_counts())
print(df['Status'].value_counts())
print(df['Gender_Focus'].value_counts())
```

```python
plt.figure()
df['Category'].value_counts().plot(kind='bar')
plt.title('Schemes by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure()
plt.scatter(df['Budget_Allocated_Crore'], df['Budget_Utilized_Crore'])
plt.title('Budget Allocated vs Utilized')
plt.xlabel('Allocated')
plt.ylabel('Utilized')
plt.show()
```

```python
plt.figure()
sns.histplot(df['Success_Rate_%'], bins=20, kde=True)
plt.title('Success Rate Distribution')
plt.show()
```

```python
top_ministry = df['Ministry'].value_counts().head(10)

plt.figure()
top_ministry.plot(kind='barh')
plt.title('Top 10 Ministries')
plt.show()
```

```python
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Avg success rate per category
print(df.groupby('Category')['Success_Rate_%'].mean())

# Budget efficiency
df['Efficiency'] = df['Budget_Utilized_Crore'] / df['Budget_Allocated_Crore']

print(df[['Scheme_Name', 'Efficiency']].head())
```

```python
plt.figure()
sns.scatterplot(x='Budget_Allocated_Crore', y='Success_Rate_%', hue='Category', data=df)
plt.title('Budget vs Success Rate')
plt.show()
```

```python
plt.figure()
sns.histplot(df['Efficiency'], bins=30, kde=True)
plt.title('Efficiency Distribution')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Category', y='Success_Rate_%', data=df)
plt.xticks(rotation=90)
plt.title('Category vs Success Rate')
plt.show()
```

```python
plt.figure()
df['Launch_Year'].value_counts().sort_index().plot()
plt.title('Schemes Launched Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Top 5 most successful schemes
top_success = df.sort_values(by='Success_Rate_%', ascending=False).head(5)
print(top_success[['Scheme_Name', 'Success_Rate_%']])
```

```python
print("Mean:", df['Success_Rate_%'].mean())
print("Median:", df['Success_Rate_%'].median())
print("Mode:", df['Success_Rate_%'].mode())
```

```python
from scipy.stats import ttest_ind

group1 = df[df['Gender_Focus'] == 'Male']['Success_Rate_%']
group2 = df[df['Gender_Focus'] == 'Female']['Success_Rate_%']

t_stat, p_val = ttest_ind(group1, group2)
print("T-test:", t_stat, p_val)
```

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df['Category'], df['Status'])
chi2, p, dof, exp = chi2_contingency(table)

print("Chi-square:", chi2, "P-value:", p)
```
